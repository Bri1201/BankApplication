```python
listSymbol = []
for i in range(len(dfnextFifty)):
    listSymbol.append(dfnextFifty.iloc[i]['Symbol'])
```

```python
ticker = yf.Ticker('ZOMATO.NS')
tickerHistory = ticker.history(interval="3mo", start="2021-06-23", end="2022-06-23")
tickerHistory
    
```

```python
nextFiftyDict = {new_list: [] for new_list in listSymbol}
```

```python
nextFiftyDict
```

```python
for i in listSymbol:
    ticker = yf.Ticker(i + '.NS')
    tickerHistory = ticker.history(interval="3mo", start="2021-05-23", end="2022-05-23")
    start = tickerHistory.iloc[0]['High']
    end = tickerHistory.iloc[2]['High']
    growth = (end - start)/start*100
    nextFiftyDict[i] = [start,end,growth]
```
