```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
#Importing data for list of top 500 companies based on market capitalisation
nseCompanies = pd.DataFrame(pd.read_excel("MCAP_31032020_TOP500.xlsx"))
nseSymbols = list(nseCompanies['Symbol'])
```

```python
nseCompanies.head()
```

```python
len(nseSymbols)
```

```python
nse = {new_list: [] for new_list in nseSymbols}
for i in nseSymbols:
    try:
        ticker = yf.Ticker(i + '.NS')
        tickerHistory = ticker.history(interval="1d", start="2022-01-14", end="2022-06-17")
        start = tickerHistory.iloc[0]['High']
        end = tickerHistory.iloc[-1]['High']
        growth = (end - start)/start*100
        nse[i] = [start,end,growth]
    except IndexError:
        nse.pop(i)
        continue
```

```python
len(nse)
```

```python
nseDf = pd.DataFrame.from_dict(nse)
```

```python
nseDf.head()
```

```python
nseFinal = nseDf.transpose()
nseFinal.head()
```

```python
nseFinal.rename(columns = {0:'1 year Ago', 1:'Today',2:'GrowthPct'}, inplace = True)
```
